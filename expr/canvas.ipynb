```python
using System;


public class ToneGenerator
{
    private double frequency;
    private double amplitude;
    private double sampleRate;

    public ToneGenerator(double frequency, double amplitude, double sampleRate)
    {
        this.frequency = frequency;
        this.amplitude = amplitude;
        this.sampleRate = sampleRate;
    }

    public double[] GenerateSamples(double duration)
    {
        int numSamples = (int)(duration * sampleRate);
        double[] samples = new double[numSamples];

        double angularFrequency = 2 * Math.PI * frequency;

        for (int i = 0; i < numSamples; i++)
        {
            double time = i / sampleRate;
            samples[i] = amplitude * Math.Sin(angularFrequency * time);
        }

        return samples;
    }
}


var tg = new ToneGenerator(440, 0.25, 44100);


// samples now contains 1 second of a 440 Hz sine wave at 25% amplitude
var excitment = tg.GenerateSamples(1);






```
